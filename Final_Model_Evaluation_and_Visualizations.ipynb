## Load The Saved Comparison Tables


```python
from pathlib import Path
import os
import pandas as pd
from IPython.display import display

limit = int(os.environ.get("FINAL_EVAL_LIMIT", "0")) or None
output_dir = Path("Final Evaluation Outputs") if limit is None else Path(f"Final Evaluation Outputs Smoke Test {limit}")
tables_dir = output_dir / "tables"
figures_dir = output_dir / "figures"

overall_metrics = pd.read_csv(tables_dir / "model_metrics_overall.csv")
source_metrics = pd.read_csv(tables_dir / "model_metrics_by_source.csv")
captions = pd.read_csv(tables_dir / "figure_captions.csv")

display(overall_metrics)
display(source_metrics)
display(captions)
```

## Display Saved Figures


```python
from IPython.display import Image, Markdown, display

for _, row in captions.iterrows():
    figure_path = figures_dir / row["figure"]
    display(Markdown(f"### {row['figure']}\n{row['caption']}"))
    display(Image(filename=str(figure_path)))
```
